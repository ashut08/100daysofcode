### Exploring pandas Library
<h3><u>Why pandas?</u></h3>
- flexibility of python
- allow to work with big data set

<h2>Import pandas Library</h2>

```python
import pandas as pd
```

<h3>Read data by read_csv method</h3>

```python
df=pd.read_csv('USA_cars_datasets.csv')
```

<h2>Reading header</h2>

```python
df.head()
```

<h3>Read each column using Slice operator</h3>

```python
df['brand'][0:4]
```

<h3> Read each row using Slice operator</h3>

```python
df.iloc[1:5]
```

### Iterating throgh each row

```python
for index,row in df.iterrows():
    print(index,row['brand'])
```

```python
df.loc[df['brand']=='jeep']
```

### Sorting Value

```python
#df.sort_values(['brand'])
#df.sort_values(['brand'],ascending=False,ignore_index=False)
df.sort_values(['brand','price'],ascending=[1,0])
```

<h2>Making some change in data</h2>

```python
df['Address']=df['state']+df['country']
df['Address']
```

```python
df.head()
```

### Droping some column

```python
df=df.drop(columns=['Address'],axis=0)
df.head()
```

### Saving our modified data using pandas

```python
df.to_csv('modified.csv')
```

```python
pd.read_csv('modified.csv')
```

<h2>Filtering Data</h2>

```python
#df.loc[df['brand']=='nissan'] 
#df.loc[(df['brand']=='nissan') & (df['model']=='versa')] 
df.loc[(df['brand']=='nissan') & (df['price']>2000)] 
```

<h2>Aggregate Statistics (Group by)</h2>

```python
df.groupby(['brand']).mean().sort_values('price')
```

```python
df.groupby(['brand']).count()
```

<h2>Load  big data with chunksize</h2>

```python
for df in pd.read_csv('modified.csv',chunksize=5):
    print(df)
```
