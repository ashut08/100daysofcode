```python
#import required library
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
import time
```

```python
chrome_options = Options()
chrome_options.add_argument('--headless') #a web browswer without GUI
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--kiosk')
chrome_options.add_argument('--ignore-certificate-errors')
chrome_options.add_argument('--incognito')
chrome_options.add_argument('--disable-dev-shm-usage')
driver=webdriver.Chrome('/home/demo/Downloads/chromedriver_linux64/chromedriver',chrome_options=chrome_options)
```

```python
driver.get('https://pluralsight.com/')
```

```python
ele=driver.find_element_by_name("q")
```

```python
ele.clear()
ele.send_keys('ashut08')
sleep(1)
ele.send_keys(Keys.RETURN)
```

```python
driver.quit()
```
