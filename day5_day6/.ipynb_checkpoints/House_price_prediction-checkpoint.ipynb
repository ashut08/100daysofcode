### Preparing data
<h5>Looking through the column we see following features contains feet('SF'):BsmtFin SF2,Bsmt Unf SF.Totoal Bsmt SF,1st fir SF, 2nd Fir SF,Low quality Fin SF</h5>
<h5>But there is no feature for total space in the house which is the sum of Basement and Upper Floors(Total Bsmt SF+Gr Liv Area) ,so lets add new feature contain the value called Total SF to the dataset



```python
#to check it have no null values
df[['Total Bsmt SF','Gr Liv Area']].isnull().values.any()
```

```python
df['Total SF']=df['Total Bsmt SF']+df['Gr Liv Area']
```

### We have to visualize the data for have idea b/w various feature

```python


# This function visualizes our data and optionally a learned line
def visualize_data(x_vals, y_vals,
                   addn_x_vals=None, addn_y_vals=None, add_addn_reg_line=False):
  
  f, ax = plt.subplots(figsize=(8,8))
  plt.plot(x_vals, y_vals, 'ro')   # red dot for each data point
  # Optionally plot another set of data points in a different color and symbol
  if (addn_x_vals is not None):
    plt.plot(addn_x_vals, addn_y_vals, 'g^') # green triangles for additional data points
    # Optionally, plot a regression line.
    if (add_addn_reg_line):
      x_min_index = addn_x_vals.argmin()
      x_max_index = addn_x_vals.argmax()
      print(x_min_index,[addn_x_vals[x_min_index],addn_y_vals[x_min_index]] ) 
      print(x_max_index,[addn_x_vals[x_max_index],addn_y_vals[x_max_index]] ) 
      plt.plot([addn_x_vals[x_min_index],addn_y_vals[x_min_index]], 
               [addn_x_vals[x_max_index],addn_y_vals[x_max_index]], 
               'b-')  # draw a blue regression line
    
  plt.tick_params(axis='both', which='major', labelsize=14)
  
  plt.show() 
```

```python
visualize_data(df['Total SF'],df['SalePrice'])
```

<h3>So visualization tell us there is in linear regression in b/w TotalSf and SalePrice</h3>

```python
from sklearn.preprocessing import MinMaxScaler
```

<h3>If values are on very different scales it will be diffcult for the model to determine the relationship between features.with our data SF rannge from 800-4200 Price Range 80,000 to 400,0000</h3>
 <h4>When we normalize we will reduce both qualities to same scale while preserving the difference b/w price and size of homes

```python
sf_scaler=MinMaxScaler()
sf_scaled=sf_scaler.fit_transform(df[['Total SF']]).reshape(-1,1).astype(np.float64)
price_scaler=MinMaxScaler()
price_scaled=price_scaler.fit_transform(df[['SalePrice']]).reshape(-1,1).astype(np.float64)
```

#  Create the model

```python
import tensorflow as tf
```

```python
# create model using the tensorflow keras library
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units=1,activation='linear',input_shape=(1,),kernel_initializer='random_uniform',bias_initializer='zeros'))
```

# compile model

```python
model.compile(loss='mean_squared_error',optimizer='sgd')
```

# Train the model

```python
from sklearn.model_selection import train_test_split
```

```python
sf_train_scaled,sf_test_scaled,price_train_scaled,price_test_scaled=train_test_split(sf_scaled,
                                                                    price_scaled,test_size=0.3,random_state=42)
```

# Train model using train data

```python
initial_epochs = 1200
batch_size = 10
train_hist = model.fit(sf_train_scaled, price_train_scaled, 
                       epochs=initial_epochs, batch_size=batch_size, verbose=1)
```

<h3>Our model learned the best weight and bias for our training data that it could through the defined number of epochs. Let's see if a line drawn with the learned weight and bias looks reasonable.</h3>

```python
price_predicted_scaled =  model.predict(sf_train_scaled)
visualize_data(sf_scaler.inverse_transform(sf_train_scaled), price_scaler.inverse_transform(price_train_scaled),
               sf_scaler.inverse_transform(sf_train_scaled), price_scaler.inverse_transform(price_predicted_scaled),
               add_addn_reg_line=False)
```

```python
def plot_loss(hist):
  # plot the loss
  plt.title('Loss History')
  plt.plot(hist.history['loss'])
  plt.ylabel('loss')
  plt.xlabel('epoch')
  plt.show()

print(type(train_hist))
plot_loss(train_hist)
```

```python
price_predicted_scaled = model.predict(sf_test_scaled)

# calculate the mean squared error for the prediction, lower is better
from sklearn.metrics import mean_squared_error
print("prediction mean squared error:", 
          mean_squared_error(price_test_scaled, price_predicted_scaled))

# Visualize data
#    plus the test square footage and predicted price
print('\nGreen triangles are Test square footage with predicted Price')
visualize_data(sf_scaler.inverse_transform(sf_test_scaled),
               price_scaler.inverse_transform(price_test_scaled),
               sf_scaler.inverse_transform(sf_test_scaled), 
               price_scaler.inverse_transform(price_predicted_scaled),
               add_addn_reg_line=False)
print('\nOur predicted values fit the data well!')

```
