```python
import requests
from bs4 import BeautifulSoup as bs
import re
```

```python
r=requests.get('https://ashut08.github.io/ashu/')
```

```python
r
```

```python
#conver to  beautiful soup object
soup=bs(r.content)
```

```python
print(soup.prettify())
```

```python
#paragraph=soup.find('h2',string=re.compile("Music"))  choose music only capital letter
paragraph=soup.find('h2',string=re.compile("(m|)Music")) # choose music with small and capital 
```

```python
print(paragraph)
```
