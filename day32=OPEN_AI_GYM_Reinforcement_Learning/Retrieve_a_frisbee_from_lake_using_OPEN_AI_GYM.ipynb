```python
import gym 
import numpy as np
```

```python
env=gym.make('FrozenLake-v0')
```

```python
#the no action possible to every state
#ther are foure possible actin left,right,top,bottom
env.action_space.n
```

```python
#its way to describe current state of envoirement
env.observation_space
```

## Its 4X4 grid with 16 tiles (S,F,G,H)
* S=start
* F=Frozen
* H=hole
* G=goal

```python
# initialize learning rate
alpha=0.4
```

```python
#initialize discount factor 
gamma=0.99
```

```python
#initialize q-table
q_table=dict([(x,[1,1,1,1]) for x in range(16)])
```

```python
q_table
```

```python
# define a helper method which choose action based on the current observation
def choose_action(observ):
    return np.argmax(q_table[observ])
```

```python
for i in range(1000):
    observ=env.reset()
    action=choose_action(observ)
    prev_observ=None
    prev_action=None
    t=0
    for t in range(2500):
        env.render()
        observ,reward,done,info=env.step(action)
        action=choose_action(observ)
        if not prev_observ is None:
            q_old=q_table[prev_observ][prev_action]
            q_new=q_old
            if done:
                q_new+=alpha*(reward-q_old)
            else:
                q_new+=alpha*(reward+gamma*q_table[observ][action]-q_old)
            new_table=q_table[prev_observ]
            new_table[prev_action]=q_new
            q_table[prev_observ]=new_table
        prev_observ=observ
        prev_action=action
        if done:
            print("Episode{} finished after{} with r{}.".format(i,t,reward))
            break
```
