```python
import sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
train_text=["The pessimist See diffculty In Every Oppertunity",
           "The Optimist See Oppertunity In Every Diffculty",
           "Don't Let Yesterday Take Up Too Much Of Today",
           "You Learn More From Failure Than From Succes"
           "This is book"]
```

```python
n_gram_vectorizer=CountVectorizer(ngram_range=(2,2))
```

```python
transformed_vector=n_gram_vectorizer.fit_transform(train_text)
```

```python
n_gram_vectorizer.vocabulary_ # here you see that all of vacablury in the form of bigram
```
