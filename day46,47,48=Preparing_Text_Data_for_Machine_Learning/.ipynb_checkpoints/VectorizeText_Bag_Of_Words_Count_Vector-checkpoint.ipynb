```python
import sklearn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
train_text=["The pessimist See diffculty In Every Oppertunity",
           "The Optimist See Oppertunity In Every Diffculty",
           "Don't Let Yesterday Take Up Too Much Of Today",
           "You Learn More From Failure Than From Succes"
           "This is book"]
```

```python
# tokenize the input text into words and convert into numeric
count_vectorizer=CountVectorizer()
```

```python
count_vectorizer.fit(train_text)
```

```python
count_vectorizer.get_feature_names() # it give vocabulary of your train text
```

```python
count_vectorizer.vocabulary_
```

```python
count_vectorizer.vocabulary_.get("see")
```

```python
#creating feature vector from training data set
transformed_vector=count_vectorizer.transform(train_text)
```

```python
transformed_vector
```

```python
print(transformed_vector.shape) # 6 sentence are featurized and 24 are unique words
```

```python
print(transformed_vector.toarray()) #conver into dense array
                                    # 1 indicate these word occur one time and 2 indicate it occurs two time
```

```python
#inverse_transform operation that allows to regenerate the original sentence
count_vectorizer.inverse_transform(transformed_vector)
```

```python

text_text=["A Room without books is like a body without soul"]
```

```python
count_vectorizer.transform(text_text).toarray()
```
