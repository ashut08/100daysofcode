```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
train_text=["The pessimist See diffculty In Every Oppertunity",
           "The Optimist See Oppertunity In Every Diffculty",
           "Don't Let Yesterday Take Up Too Much Of Today",
           "You Learn More From Failure Than From Succes"
           "This is book"]
```

```python
tfidf_vectorizer=TfidfVectorizer()
```

```python
transformed_vector=tfidf_vectorizer.fit_transform(train_text)
```

```python
tfidf_vectorizer.vocabulary_
```

```python
transformed_vector.toarray()
```

```python
transformed_vector.toarray().shape
```

```python
tfidf_vectorizer.idf_ # word occurs very commanly across your document will have low idf
                      # words that are rare in your document have higher
```

```python
#now we check which word have higher and lower idf score
list(zip(tfidf_vectorizer.get_feature_names(),tfidf_vectorizer.idf_))
```
